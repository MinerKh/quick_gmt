```python
import pygmt
import numpy as np
```

```python
def plot_special_latitudes(fig, lons):
    """
    在地图上绘制北回归线、南回归线、北极圈、南极圈
    """
    special_lines = [
        {"lat": 23.5, "pen": "0.5p,gray20,--", "label": "Tropic of Cancer"},
        {"lat": -23.5, "pen": "0.5p,gray20,--", "label": "Tropic of Capricorn"},
        {"lat": 66.5, "pen": "0.5p,gray20,--", "label": "Arctic Circle"},
        {"lat": -66.5, "pen": "0.5p,gray20,--", "label": "Antarctic Circle"},
    ]
    for line in special_lines:
        fig.plot(
            x=lons,
            y=np.full_like(lons, line["lat"]),
            pen=line["pen"],
            label=line["label"]
        )
```

```python
# 创建地图
fig = pygmt.Figure()
# Robinson投影，中心经线0度
# 绘制海洋
fig.coast(
    region="g",
    water="skyblue@50",  # 海洋颜色，50%透明
    projection="N0/15c",
)
fig.show()
```

```python
fig.basemap(
    frame="a30f15g15",
    # map_scale="jBR+o1c/1c+w2000k+f+lkm+ar",
    # box="+gwhite@30+p0.5p,gray30,solid+r3p"
    )
fig.show()
```

海岸线分辨率可选项f、h、i、l、c  
**f**ull > **h**igh > **i**ntermediate > **l**ow > **c**rude

```python
# 绘制陆地
fig.coast(
    land="white",        # 大陆颜色改为白色
    borders=[1],      # 1:国界 2:省界
    shorelines="1/0.01p,black",  # 海岸线
    resolution="c", #较低的分辨率
)
fig.show()
```

绘制五带

```python
# 生成经度数组
lons = np.linspace(-180, 180, 361)
plot_special_latitudes(fig, lons)
fig.show()
```

```python
# 添加美观的比例尺，右下角
fig.basemap(map_scale="n0.7/0.2+c+w4000k+f+l")
# 添加指北针（显示N、E、S、W）
fig.basemap(compass="jBC+w1.5c+o0c/0.5c+lW,E,S,N")
fig.show()
```
