导入pygmt库

```python
import pygmt
```

设置绘图区域，以东北亚地区为例。  
从全球地形起伏数据earth——relief 以60分到1秒的分辨率加载数据。  
分辨率resolution，全球地形先用30m、全国地形起伏用05m、3度X3度用30s、更小的区域用03s。

```python
region = [110, 155, 30, 55]
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=region)
```

创建图片，设置基本参数。

```python
fig = pygmt.Figure()
fig.grdview(
    grid=grid,
    # 平面视角和图像的高度角
    perspective=[180, 90],
    # 图像配色
    cmap="geo",
    frame=["xaf", "yaf", "WSNE"],
    # 墨卡托投影，图片长15cm
    projection="M15c",
    zsize="1.5c",
    surftype="s",
)
fig.show()
```

```python
# 添加高程比例尺 比例尺的间隔是1000m
fig.colorbar(perspective=True, frame=["a1000", "x+lElevation", "y+lm"])

# 显示图像
fig.show()
```

保存图片

```python
fig.savefig("CN_NE_GMT.png", dpi=600)
```
